```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Dataset .csv.zip')

df.info()
df.head()
```

```python
#top 10 cuisines
top_cuisines = df["Cuisines"].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="coolwarm")
plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine")
plt.show()
```

```python
#price range distribution
plt.figure(figsize=(8, 5))
sns.countplot(x="Price range", data=df, palette="viridis")
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# Boxplot of Average Cost for Two by Price Range
plt.figure(figsize=(10, 5))
sns.boxplot(x="Price range", y="Average Cost for two", data=df, palette="muted")
plt.ylim(0, 5000)  #Limit
plt.title("Average Cost for Two Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Average Cost for Two")
plt.show()
```

```python
# Relationship between votes and ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Votes", y="Aggregate rating", data=df, alpha=0.5)
plt.title("Votes vs. Aggregate Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
plt.show()
```

# **Insights from Dataset**
- The majority of restaurants receive ratings between 3.0 and 4.5, with fewer establishments falling at the extreme ends of the rating scale.  
- The **top 10 most popular cuisines** reveal clear consumer preferences, shaping the market landscape.  
- The **price range distribution** highlights that **mid-range dining options** dominate the market, catering to a broad customer base.  
- **Higher-end restaurants** exhibit substantial variation in pricing, reflecting differences in ambiance, service, and exclusivity.  
- A **strong correlation exists between the number of votes and ratings**, suggesting that higher engagement often aligns with better customer satisfaction.  
- **Dining costs for two people** fluctuate significantly, even within the same price category, indicating varied offerings within each segment.  
- Some **cuisine types vastly outnumber others**, emphasizing their popularity and demand among diners.  
- A large proportion of restaurants **do not provide online delivery**, suggesting a preference for **traditional dine-in experiences** over digital convenience.  
- **Better-rated restaurants** generally attract more votes, highlighting the role of customer engagement in influencing restaurant reputation.  
- Certain cuisines appear to be **more closely linked with higher ratings**, potentially reflecting customer expectations and culinary excellence.
